```python

#setup the system of equation
equations = [
    Eq(dS_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dE_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dIs_dt.subs({E:0,Is:0,Ia:0,B:0,H:0}),0),
    Eq(dIa_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dH_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dR_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dW_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dV1_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dV2_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dV3_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dVB_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
    Eq(dB_dt.subs({E:0,Is:0,Ia:0,B:0, H:0}),0),
]
```
